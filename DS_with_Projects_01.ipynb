### 모델 검증
All Data
- Training Data
- Test Data

`Train Data`를 통해 학습한 모델을 `Test Data`를 통해 검증하는데, 모델이 얼마나 잘 `Fit`을 했는지 보기 위해 `Cross Validation`을 하여 `Data`를 나눠주게 된다.

예:

`All Data`를 5개로 나누고 이를 `Fold`라고 한다.
`Fold1, Fold2, Fold3, Fold4, Fold5`
그리고, 이렇게 `Fold`를 나눠준 수만큼 `Fit`을 한다. 
이때, 각 `Fold`가 하나씩 돌아가면서 `Test Data`의 역할을 하고, 나머지가 `Train Data`의 역할을 하게 된다.
`Split1, Split2, Split3, Split4, Split5`
**각 `Fold`마다 점수를 평균을 내고, 가장 좋은 점수를 낸 모델을 채택하고, 최적의 파라미터를 얻는다.**

예:

`clf = SVC()`
`clf.fit(X_train, y_train)`
`y_pred = clf.predict(X_test)`
-> `SVC()` 안에 여러가지 파라미터값이 들어가는데, 어떤 파라미터가 들어가면 좋을지 하나하나 테스트하기가 어려움.
-> 그렇기 때문에 `Cross Validation`을 통해 여러번 학습하여, **파라미터 값을 찾는 것**

### Overfitting and Underfitting
당연하게도 학습을 적게 하면 그만큼 정확도도 낮다.

**학습을 적게 하여 정확도가 낮은 상태를 Underfitting 이라고 한다.**

그렇다고 학습을 많이 하면 많이 할수록 무한하게 정확도를 올릴 수 있는 것은 아니다.

Train 데이터 학습을 많이 하면 할수록 Train 데이터에 대한 정확도는 올라가지만 Test 데이터에 대한 **정확도는 올라가다가 어느 수준에 다다르면 그 이후부터는 오히려 감소하는 경향이 나타난다.**

**이를 Overfitting 되었다고 한다.**

**Overfitting이 되기 전 모델이 최적의 성능을 보여주는 `Sweet spot`을 발굴하는 것이 매우 중요하다.**

### 분류(Classification)

```python
from sklearn import tree
X = [[0,0], [1,1]]
Y = [0,1]
```

```python
clf = tree.DecisionTreeClassifier()
clf
```

```python
# 학습(Fit)
clf = clf.fit(X, Y)
clf
```

```python
# iris dataset을 활용한 모델 예시
from sklearn.datasets import load_iris
from sklearn import tree
X, y = load_iris(return_X_y=True)
X, y
```

iris dataset은 숫자로만 이루어져있지 않은데 숫자 배열로만 출력되는 이유는 머신러닝에 해당 dataset을 학습시키기 위한 것이며, 0, 1, 2는 단순 숫자 배열로 보이지만 일종의 카테고리라고 생각하면 된다.

```python
# 학습(Fit)
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
# 시각화
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 10))
t = tree.plot_tree(clf.fit(X, y), filled=True)
```

의사결정나무는 스무고개와 같다.

```python
# graphviz를 활용한 시각화
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render('iris')
```

```python
dot_data = tree.export_graphviz(clf, out_file=None,
                                feature_names=iris.feature_names,
                                class_names=iris.target_names,
                                filled=True, rounded=True,
                                special_characters=True)
graph = graphviz.Source(dot_data)
graph
```

### Pima indian dataset
**서구화로 인한 Pima indian들의 당뇨병 발병에 대한 dataset[1]**

**변수(9columns)**
- Pregnancies : 임신 횟수
- Glucose : 2시간 동안의 경구 포도당 내성 검사에서 혈장 포도당 농도
- BloodPressure : 이완기 혈압 (mm Hg)
- SkinThickness : 삼두근 피부 주름 두께 (mm), 체지방을 추정하는데 사용되는 값
- Insulin : 2시간 혈청 인슐린 (mu U / ml)
- BMI : 체질량 지수 (체중kg / 키(m)^2)
- DiabetesPedigreeFunction : 당뇨병 혈통 기능
- Age : 나이
- **Outcome(예측 타겟값)** : 768개 중에 268개의 결과 클래스 변수(0 또는 1)는 1이고 나머지는 0입니다.

[1]:https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database

### 필요한 라이브러리 로드

```python
# pandas(데이터분석)
# numpy(수치계산)
# seaborn, matplotlib.pyplot(시각화)

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('data/diabetes.csv')
df.shape
```

```python
df.head()
```

### 학습, 예측 dataset 나누기

```python
# 8:2 비율
# 전체 데이터의 행에서 80% 위치에 해당하는 index 값을 split_count 변수에 담아주기

split_count = int(df.shape[0] * 0.8)
split_count
```

```python
# train, test로 슬라이싱을 통해 데이터 나누기

train = df[:split_count].copy()
train.shape
```

```python
test = df[split_count:].copy()
test.shape
```

### 학습, 예측에 사용할 변수(컬럼)

```python
# feature_names 라는 변수에 학습과 예측에 사용할 컬럼명을 가져오기
# -1까지 슬라이싱한 이유는 마지막 Outcome은 예측할 타겟값이기 때문에 제외하기 위해

feature_names = train.columns[:-1].tolist()
feature_names
```

### 정답값이자 예측해야 될 값

```python
# label_name 이라는 변수에 예측할 컬럼 이름을 담기

label_name = train.columns[-1]
label_name
```

❓ **feature_names는 list, label_name은 str인 이유**
- 여러 개의 변수(feature_names)를 활용하여 단일 변수(label_name)의 값을 예측하는 것이기 때문
- 간단하게, 여러 개와 한 개의 차이

### 학습, 예측 dataset 만들기

```python
# 학습 세트 만들기 (기출)

X_train = train[feature_names]
print(X_train.shape)
X_train.head()
```

```python
# 정답 값을 만들기 (기출의 정답)

y_train = train[label_name]
print(y_train.shape)
y_train.head()
```

X_train(기출, 행렬 형태), y_train(기출의 정답, 벡터 형태)

❗️ 기출 행의 수, 정답 행의 수는 같음

```python
len(X_train) == len(y_train)
```

```python
# 예측 세트 만들기 (시험)

X_test = test[feature_names]
print(X_test.shape)
X_test.head()
```

```python
# 예측의 정답값을 만들기 (시험의 정답)

y_test = test[label_name]
print(y_test.shape)
y_test.head()
```

```python
len(X_test) == len(y_test)
```

실전에서는 정답을 알 수 없지만, 모델의 성능 측정을 위한 정답 값을 만들어 준 것 

이때, test와 train의 컬럼은 `같은 수`에 `같은 것` 이어야함

### 머신러닝 알고리즘 가져오기

```python
# 당뇨병인지 아닌지 여부 판별 문제는 분류 Classifier

from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model
```

### 학습(훈련)
- 시험을 보기 전, 기출(X_train)과 정답(y_train)을 보고 공부하는 과정과 유사함

```python
model.fit(X_train, y_train)
```

### 예측
- 실전 시험(X_test)라고 보면 됨
- 우리가 정답을 직접 예측하는 과정

```python
y_predict = model.predict(X_test)
y_predict[:5]
```

실전 시험이기 때문에 X(문제 parameter)만 들어간 것이다.(정답은 추후에 예측값과 비교)

### 트리 알고리즘 분석하기
- 의사결정나무 시각화

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20, 20))
tree = plot_tree(model,
                feature_names=feature_names,
                filled=True,
                fontsize=10)
```

가장 위(root)에 위치한 **Glucose**가 Outcome에 영향을 주는 가장 중요한 변수임을 확인할 수 있음

`plot_tree`로 시각화가 가능하지만, 겹쳐서 보이거나 작게 보이기 때문에 직관적이지 않음

```python
# graphviz로 시각화

import graphviz
from sklearn.tree import export_graphviz

dot_tree = export_graphviz(model,
                          feature_names=feature_names,
                          filled=True)
graphviz.Source(dot_tree)
```

**Overfitting**
- Tree의 가지가 내려갈수록 데이터가 train 데이터에 집중해서 학습하는 경향이 나타남
- 이를 방지하기 위해 트리의 깊이를 제한하기도 함

```python
# 피처의 중요도 추출하기
# 우리가 위에서 지정한 feature의 순서대로 중요도가 출력됨

model.feature_importances_
```

```python
# 피처의 중요도 시각화 하기
# Glucose가 당뇨병을 예측하는 데에 가장 중요한 변수임을 알 수 있음
# 인슐린이 의외로 중요도가 낮은 것으로 나옴(추후에 이 중요도가 유의미한지 검정해볼 수 있음)

sns.barplot(x=model.feature_importances_, y=feature_names);
```

### 정확도(Accuracy) 측정하기

```python
# 예측값 갯수 구하기
# 실제값 - 예측값을 빼주면 같은 값은 0으로 나오게 됨
# 여기에서 절대값을 씌운 값이 1인 값이 다르게 예측한 값이 됨
# sum()을 통해 다르게 예측한 값의 갯수를 확인할 수 있음 (38개)

diff_count = abs(y_test - y_predict).sum()
diff_count
```

❗️ Tree 모델은 돌릴 때마다 예측값이 변하기 때문에 diff_count 또한 고정된 것은 아니다.

```python
# 예측의 정확도를 구하기
# 100점 만점 중에 몇 점을 맞았는지 구하는 것

(len(y_test) - diff_count) / len(y_test) * 100
```

```python
# 알고리즘을 통해 구하기

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict) * 100
```

```python
# model의 score 메서드로 구하기
# 정답값을 알고 있을 때만 계산 가능

model.score(X_test, y_test) * 100
```

이번 실습에서는 아무런 전처리를 하지 않고, 8:2 비율로 train과 test를 나누고 의사결정 나무를 돌려보고 정확도 약 75%를 얻어내었다.
